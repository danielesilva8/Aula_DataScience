```python
import matplotlib.pyplot as plt
import pandas as pd

# Leitura do CSV
ocorrencias = pd.read_csv('ocorrencias_sac.csv', names=['ocorrencia', 'data_ocorrencia'])

# Alterar a data para o formato datetime,
# se houver valores que não podem ser convertidos para datetime, eles serão substituídos por NaT (not a time)
ocorrencias['data_ocorrencia'] = pd.to_datetime(ocorrencias['data_ocorrencia'], errors='coerce')

# Contar as ocorrências por dia - value_counts = conta quantas vezes a data aparece na coluna data_ocorrencia
# .sort_index ordenar a série pelo índice (datas) em ordem crescente
ocorrencias_por_dia = ocorrencias['data_ocorrencia'].value_counts().sort_index()

# Calcular a média e o desvio padrão
# mean = calcula a média das contagens por dia
# std = calcula o desvio padrão das contagens por dia
media = ocorrencias_por_dia.mean()
desvio_padrao = ocorrencias_por_dia.std()

# Plotar o histograma das contagens de ocorrências
# figsize = cria uma nova figura para o gráfico com um tamanho de 10 polegadas de largura por 5 de altura
# bins=10 cria intervalos de 10 para o histograma
# color = define a cor das barras
# alpha=0.7 define a transparencia das barras
# edgecolor = define a cor da borda das barras do histograma
plt.figure(figsize=(10, 5))
plt.hist(ocorrencias_por_dia, bins=10, color='blue', alpha=0.7, edgecolor='black')

# Adicionar linhas para a média e o desvio padrão
# plt.axvline = adiciona uma linha vertical no gráfico na posição da média das contagens
# plt.axvline (+) = adiciona uma linha vertical no gráfico na posição da média mais o desvio padrão
# plt.axvline (-) = adiciona uma linha vertical no gráfico na posição da média menos o desvio padrão
# Essas linhas ajudam a visualizar onde a média e os devsios padrão estão localizados em relação à distribuição das ocorrências
plt.axvline(media, color='red', linestyle='--', label='Média')
plt.axvline(media + desvio_padrao, color='green', linestyle='--', label='Média + Desvio Padrão')
plt.axvline(media - desvio_padrao, color='orange', linestyle='--', label='Média - Desvio Padrão')

# Configurações do gráfico
plt.title('Histograma da Distribuição de Ocorrências no SAC')
plt.xlabel('Número de Ocorrências por Dia')
plt.ylabel('Frequência')
plt.legend()
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
