```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# cria dois arrays das funções seno e conseno
x = np.arange(0,3 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('eixo x')
plt.title('Seno e cosseno')
plt.legend(['Seno', 'Cosseno'])
plt.show()
```

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.plot(x,y_sin)
plt.title('Seno')

plt.subplot(1,2,2)
plt.plot(x,y_cos)
plt.title('Coseno')

plt.show()


```

```python
plt.plot(x, y_cos, color='tab:red', linestyle='--',marker='s')
plt.show();

plt.plot(x, y_cos,'m:', drawstyle='steps-post')
plt.show();

x = np.arange(x.size)
plt.bar(x, y_cos);
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

Dados = np.random.randint(0,10,50) # random, sempre que rodar vai aparecer em um formato diferente

plt.hist(Dados,bins=10, rwidth=0.9)
plt.show()
```

```python
x = np.random.normal(-1, 1, 1000)
y = np.random.normal(-1, 1, 1000)

fig = plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.scatter(x, y) # distribuição mais poluída
plt.subplot(1,2,2); plt.scatter(x, y, alpha=0.25);plt.show(); # parametro alpha suavizar o evento de cada ponto
```
